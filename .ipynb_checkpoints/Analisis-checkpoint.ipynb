PANDAS - Analisis de Informacion de Estudiantes

1. Cargar el Dataset
Objetivo: Cargar los datos en un DataFrame utilizando pandas.
Importancia: Esto permite trabajar con los datos de manera estructurada y explorar su contenido.
2. Exploración Básica de Datos
Se verificaron valores nulos (isnull().sum()), se obtuvo un resumen del dataset (df.info()), y se calcularon estadísticas descriptivas (df.describe()).
Interpretación:
La ausencia de valores nulos indica que no se necesita limpieza adicional.
Las estadísticas como la media y desviación estándar ayudan a entender la dispersión y distribución de las calificaciones.
3. Análisis Exploratorio de Datos (EDA)
Se utilizaron gráficos de caja y de barras para identificar patrones en las calificaciones:

Distribución por género:

Los gráficos de caja mostraron cómo las calificaciones varían según el género. Por ejemplo, se podría observar si un género sobresale en matemáticas mientras el otro lo hace en escritura.
Nivel educativo de los padres:

Un gráfico de barras indicó que los estudiantes con padres que tienen mayor nivel educativo tienden a obtener mejores promedios.
Tipo de almuerzo y curso de preparación:

Los gráficos de caja mostraron cómo estas variables influyen en las calificaciones. Por ejemplo, los estudiantes que toman almuerzo estándar o completan el curso de preparación suelen tener mejores calificaciones.
4. Análisis de Correlación
Objetivo: Entender la relación entre variables como las calificaciones en diferentes materias.
Herramienta: Un mapa de calor mostró una fuerte correlación positiva entre las calificaciones de matemáticas, lectura y escritura.
5. Análisis Estadístico
Se realizó un T-Test para comparar calificaciones de matemáticas entre géneros, y un ANOVA para evaluar diferencias en promedios según el nivel educativo de los padres.
Interpretación:
Los valores P indican si las diferencias observadas son estadísticamente significativas.

```python
# Importar las bibliotecas necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# ===========================
# 1. Cargar el Dataset
# ===========================
# Cargar el archivo CSV en un DataFrame
df = pd.read_csv("Cleaned_Students_Performance.csv")  # Ajusta la ruta si es necesario

# Mostrar las primeras filas del dataset
print("Primeras filas del dataset:")
print(df.head())

# ===========================
# 2. Exploración Básica de Datos
# ===========================
# Verificar valores nulos
print("\nValores nulos por columna:")
print(df.isnull().sum())

# Obtener información general del dataset
print("\nInformación del dataset:")
df.info()

# Obtener estadísticas descriptivas para las columnas numéricas
print("\nEstadísticas descriptivas:")
print(df.describe())

# ===========================
# 3. Análisis Exploratorio de Datos (EDA)
# ===========================

# Estilo de gráficos
sns.set(style="whitegrid")

# ---------------------------
# Distribución de Matemáticas por Género
# ---------------------------
plt.figure(figsize=(10, 6))
sns.boxplot(x='gender', y='math_score', data=df, palette="Blues", hue="gender")
plt.title('Calificaciones de Matemáticas por Género')
plt.xlabel('Género')
plt.ylabel('Calificación de Matemáticas')
plt.show()

# ---------------------------
# Distribución de Lectura por Género
# ---------------------------
plt.figure(figsize=(10, 6))
sns.boxplot(x='gender', y='reading_score', data=df, palette="Greens", hue="gender")
plt.title('Calificaciones de Lectura por Género')
plt.xlabel('Género')
plt.ylabel('Calificación de Lectura')
plt.show()

# ---------------------------
# Distribución de Escritura por Género
# ---------------------------
plt.figure(figsize=(10, 6))
sns.boxplot(x='gender', y='writing_score', data=df, palette="Oranges", hue="gender")
plt.title('Calificaciones de Escritura por Género')
plt.xlabel('Género')
plt.ylabel('Calificación de Escritura')
plt.show()

# ---------------------------
# Promedios por Nivel Educativo de los Padres
# ---------------------------
df['average_score'] = df[['math_score', 'reading_score', 'writing_score']].mean(axis=1)

plt.figure(figsize=(12, 6))
sns.barplot(x='parental_level_of_education', 
            y='average_score', 
            data=df, 
            hue='parental_level_of_education', 
            palette="coolwarm", 
            dodge=False, 
            legend=False)
plt.title('Promedios por Nivel Educativo de los Padres')
plt.xlabel('Nivel Educativo de los Padres')
plt.ylabel('Promedio de Calificaciones')
plt.xticks(rotation=45)
plt.show()

# ---------------------------
# Impacto del Tipo de Almuerzo
# ---------------------------
plt.figure(figsize=(10, 6))
sns.boxplot(x='lunch', y='average_score', data=df, palette="Set3", hue="lunch")
plt.title('Promedio de Calificaciones por Tipo de Almuerzo')
plt.xlabel('Tipo de Almuerzo (1 = Estándar, 0 = Gratis/Reducido)')
plt.ylabel('Promedio de Calificaciones')
plt.show()

# ---------------------------
# Impacto del Curso de Preparación para Exámenes
# ---------------------------
plt.figure(figsize=(10, 6))
sns.boxplot(x='test_preparation_course', y='average_score', data=df, palette="Set2", hue="test_preparation_course")
plt.title('Promedio de Calificaciones por Curso de Preparación')
plt.xlabel('Curso de Preparación (1 = Completado, 0 = No Completado)')
plt.ylabel('Promedio de Calificaciones')
plt.show()

# ===========================
# 4. Análisis de Correlación
# ===========================

# Verificar las columnas disponibles
print("Columnas disponibles en el DataFrame:")
print(df.columns)

# Asegurarse de que no haya valores nulos en las columnas necesarias
df = df.dropna(subset=['math_score', 'reading_score', 'writing_score', 'average_score'])

# Calcular la matriz de correlación
correlation_matrix = df[['math_score', 'reading_score', 'writing_score', 'average_score']].corr()

# Mostrar la matriz de correlación
print("\nMatriz de correlación:")
print(correlation_matrix)

# Crear el mapa de calor
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='bwr', linewidths=0.5)
plt.title('Matriz de Correlación de las Calificaciones')
plt.show()

# ===========================
# 5. Análisis Estadístico
# ===========================

# ---------------------------
# T-Test: Comparación de Matemáticas por Género
# ---------------------------
# Verificar valores únicos en 'gender'
print("\nValores únicos en 'gender':", df['gender'].unique())

# Ajustar los filtros según los valores correctos
male_math_scores = df[df['gender'] == 'male']['math_score']
female_math_scores = df[df['gender'] == 'female']['math_score']

# Eliminar valores nulos
male_math_scores = male_math_scores.dropna()
female_math_scores = female_math_scores.dropna()

# Verificar tamaños de muestra
print("\nTamaño del grupo masculino:", len(male_math_scores))
print("Tamaño del grupo femenino:", len(female_math_scores))

# Realizar el T-Test si hay suficientes datos
if len(male_math_scores) > 1 and len(female_math_scores) > 1:
    t_statistic, p_value = stats.ttest_ind(male_math_scores, female_math_scores)
    print(f"\nT-test: T-statistic = {t_statistic}, P-value = {p_value}")
else:
    print("\nNo hay suficientes datos para realizar el T-Test.")

# ---------------------------
# ANOVA: Promedios por Nivel Educativo de los Padres
# ---------------------------
anova_result = stats.f_oneway(
    df[df['parental_level_of_education'] == "high school"]['average_score'],
    df[df['parental_level_of_education'] == "associate's degree"]['average_score'],
    df[df['parental_level_of_education'] == "bachelor's degree"]['average_score'],
    df[df['parental_level_of_education'] == "master's degree"]['average_score'],
    df[df['parental_level_of_education'] == "some college"]['average_score'],
    df[df['parental_level_of_education'] == "some high school"]['average_score']
)

print(f"\nANOVA: F-statistic = {anova_result.statistic}, P-value = {anova_result.pvalue}")
```

sns.boxplot

La función sns.boxplot crea gráficos de caja que muestran la distribución de una variable continua a través de categorías.

Parámetros:
x: Variable categórica (por ejemplo, género).
y: Variable continua (por ejemplo, calificaciones).
data: DataFrame que contiene las variables.
hue: Variable para dividir las categorías en subgrupos (por ejemplo, género dentro de cada nivel educativo).
palette: Paleta de colores para los gráficos.
Funcionamiento:
Crea un resumen visual de la distribución de datos:
El rectángulo central muestra el rango intercuartílico (Q1 a Q3).
La línea dentro del rectángulo indica la mediana.
Los puntos fuera de los "bigotes" representan valores atípicos.
Compara categorías fácilmente:
Por ejemplo, comparar calificaciones entre géneros o tipos de almuerzo.

Ejemplo en el análisis:

```python
sns.boxplot(x='gender', y='math_score', data=df, palette="Blues", hue="gender")
```

sns.heatmap

La función sns.heatmap crea mapas de calor para visualizar relaciones entre variables, como una matriz de correlación.

Parámetros:
data: Matriz (por ejemplo, una matriz de correlación) a visualizar.
annot: Muestra los valores numéricos en cada celda.
cmap: Esquema de colores para los valores.
vmin y vmax: Establecen los límites mínimo y máximo del mapa.
linewidths: Tamaño de las líneas entre las celdas.
Funcionamiento:
Traduce datos numéricos en colores:
Los valores altos (por ejemplo, correlaciones fuertes) tienen colores más intensos.
Los valores bajos o negativos tienen colores más suaves o diferentes tonalidades.
Identifica patrones fácilmente:
Por ejemplo, relaciones fuertes entre variables.

Ejemplo en el análisis:

```python
sns.heatmap(correlation_matrix, annot=True, cmap='bwr', linewidths=0.5)
```
